```python
import pandas as pd 
import requests
import os
import zipfile
```

```python
anos = range(2010, 2024)

url_base = f'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/'

os.chdir(r'C:\Users\Rafael\dev\dados_cvm')
```

```python
for ano in anos:

       download = requests.get(url_base + f'DFP/DADOS/dfp_cia_aberta_{ano}.zip')

       open(f"dfp_cia_aberta_{ano}.zip", "wb").write(download.content)
```

```python
lista_demonstracoes_2010_2023 = []
```

```python
diretorio_atual = os.getcwd()


for arquivo in os.listdir(diretorio_atual):

    arquivo_zip = zipfile.ZipFile(os.path.join(diretorio_atual, arquivo))

    tipo = "dfp"

    ano = arquivo[-8: -4]

    for planilha in arquivo_zip.namelist():

        if planilha != f"{tipo}_cia_aberta_{ano}.csv":

            demonstracao = pd.read_csv(arquivo_zip.open(planilha), sep= ";", decimal = ",", encoding = 'ISO-8859-1', chunksize=1000)

            demonstracao = pd.concat(demonstracao, ignore_index=True)

            demonstracao["tipo_doc"] = len(demonstracao) * [tipo]

            lista_demonstracoes_2010_2023.append(demonstracao)

  
    
```

```python
base_dados = pd.concat(lista_demonstracoes_2010_2023)

base_dados
```

```python
base_dados[["con_ind", "tipo_dem"]] = base_dados['GRUPO_DFP'].str.split("-", expand= True)

base_dados["tipo_dem"] = base_dados["tipo_dem"].str.strip()
base_dados["con_ind"] = base_dados["con_ind"].str.strip()
base_dados["con_ind"] = base_dados["con_ind"].astype(str)

base_dados
```
