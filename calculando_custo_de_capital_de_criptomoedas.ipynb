```python
import pandas as pd 
import numpy as np 
import yfinance as yf 
from cryptocmd import CmcScraper
import statsmodels.api as sm 
```

```python
scraper = CmcScraper("BTC")
df = scraper.get_dataframe()
df = df.set_index("Date")
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')

df
```

```python
tbond10 = yf.download('IEF', "2017-01-01", "2023-12-31")['Adj Close']
tbond10 = tbond10.resample("Y").last().pct_change().dropna()
tbond10
```

```python
df_bitcoin_anual = df.resample("Y").last()['Close'].pct_change().dropna()
df_bitcoin_anual = df_bitcoin_anual.loc[df_bitcoin_anual.index > pd.Timestamp("2017-12-31")]
df_bitcoin_anual
```

```python
premio_risco = df_bitcoin_anual - tbond10
premio_risco.mean()
```

```python
retorno_esperado = tbond10.loc['2023-12-31'] + 1 *(premio_risco.mean() - tbond10.loc['2023-12-31'])

retorno_esperado
```

```python
scraper = CmcScraper("ETH")
df_eth = scraper.get_dataframe()
df_eth = df_eth.set_index("Date")

df_eth
```

```python
df_btc = df.loc[df.index >= "2015-08-07"]
df_eth = df_eth[df_eth.index >= "2015-08-07"]
```

```python
retorno_eth = df_eth['Close'].pct_change().dropna()
retorno_btc = df_btc['Close'].pct_change().dropna()
```

```python
X = retorno_btc
Y = retorno_eth
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()

print(model.params[1])
print(model.rsquared)
```
